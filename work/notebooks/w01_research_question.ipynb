## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

#### Main Question

* Which pages should be prioritized for content review or refresh?

 #### Decision Being Improved

* The decision is determining which pages a content editor should spend time reviewing first.

#### Action

* A content editor or SEO specialist would use the ranked recommendations to decide whether a page should be updated, expanded, monitored, or left unchanged.

#### Cost of a Wrong Recommendation

* A poor recommendation can waste valuable time on pages that do not need attention. On the other hand, failing to identify an important declining page could result in lost traffic and missed opportunities for improvement.

#### Why Data and ML Can Help

* Large websites often contain hundreds or thousands of pages, making manual review difficult. By analyzing performance signals such as impressions, CTR, engagement, content age, and trends, data can help prioritize pages that appear most likely to benefit from further review.

```python

print(f"Total pages: {total_pages:,}")
print(f"Declining pages: {declining:,} ({declining/total_pages:.1%})")
print(f"Declining pages with 100+ impressions: {declining_demand:,} ({declining_demand/total_pages:.1%})")
print(f"Baseline Precision@50: {base:.3f}")
print(f"Random Forest Precision@50: {rf:.3f}")
```
